题目1:两数之和暴力解法

```python
def twoSum(nums, target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i]+nums[j]==target:
                return [i,j]
    return []
    # 测试
if __name__ == "__main__":
    nums1 = [2, 7, 11, 15]
    target1 = 9
    print(f"输入: nums = {nums1}, target = {target1}")
    print(f"输出: {twoSum(nums1, target1)}")
    print(f"期望: [0, 1]\n")
```

题目2: 寻找数组中心索引

```python
def pivotIndex(nums):
    total_sum=sum(nums)
    left_sum=0
    for i,num in  enumerate(nums):
        right_sum=total_sum-left_sum-num
        if right_sum==left_sum:
            return i
        left_sum+=num
    return -1
# 测试用例
if __name__ == "__main__":
    # 测试用例1
    nums1 = [1, 7, 3, 6, 5, 6]
    print(f"输入: nums = {nums1}")
    print(f"输出: {pivotIndex(nums1)}")
    print(f"期望: 3\n")
    
    # 测试用例2
    nums2 = [1, 2, 3]
    print(f"输入: nums = {nums2}")
    print(f"输出: {pivotIndex(nums2)}")
    print(f"期望: -1\n")
    
    # 测试用例3
    nums3 = [1,-1,2,-2]
    print(f"输入: nums = {nums3}")
    print(f"输出: {pivotIndex(nums3)}")
    print(f"期望: -1\n")
```

题目3: 移动零

```python
def moveZeroes(nums):
    slow =0
    for fast in range(len(nums)):
        if nums[fast]!=0:
            nums[slow],nums[fast]=nums[fast],nums[slow]
            slow+=1
# 测试1:混合0和非0
nums1 = [0,1,0,3,12]
moveZeroes(nums1)
print(nums1)  # [1, 3, 12, 0, 0]

# 测试2:只有0
nums2 = [0]
moveZeroes(nums2)
print(nums2)  # [0]

# 测试3:没有0
nums3 = [1,2,3]
moveZeroes(nums3)
print(nums3)  # [1, 2, 3]

# 测试4:0在开头
nums4 = [0,0,1]
moveZeroes(nums4)
print(nums4)  # [1, 0, 0]

# 测试5:全是0
nums5 = [0,0,0]
moveZeroes(nums5)
print(nums5)  # [0, 0, 0]
```
